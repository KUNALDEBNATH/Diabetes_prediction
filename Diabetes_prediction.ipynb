```python
#import the libraries or dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler #to standerize data
from sklearn.model_selection import train_test_split
from sklearn import svm #model
from sklearn.metrics import accuracy_score
```

```python
#Data collection and analysis
diabetes_dataset = pd.read_csv('/content/large_diabetes_prediction_dataset.csv', header=0)
```

```python
diabetes_dataset.head()
```

```python
#no. of rows and col
diabetes_dataset.shape
```

```python
#getting statistical measure of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['diabetes'].value_counts()
```

```python
#0--> Non-diabetic
#1--> Diabetic

# Select only numeric columns before calculating the mean
numeric_features = diabetes_dataset.select_dtypes(include=['number'])

diabetes_dataset.groupby('diabetes')[numeric_features.columns].mean()
```

```python
X=diabetes_dataset.drop(columns='diabetes',axis=1)
Y=diabetes_dataset['diabetes']
```

```python
print(X)
print(Y)
```

```python
#Data Standerization
scaler = StandardScaler()
```

```python
# Assuming X is your DataFrame
X = X.select_dtypes(include=['number'])  # Select only numeric columns

# Now apply StandardScaler
k = scaler.fit_transform(X)
```

```python
X=k
print(X.shape)
y=diabetes_dataset['diabetes']
print(X)
print(Y)
```

```python
#Train Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
#Training the model    SVC-SUPPORT VECTOR CLASSIFIER   KERNEL--> PARAMETER
classifier = svm.SVC(kernel='linear')
```

```python
#training the svm classifier
classifier.fit(X_train,Y_train)
```

```python
# accuracy score on the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train) #comparing X with Y
```

```python
print(training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```

```python
#making a predictive system
input_data = (8,125,96,0,0,0,0.232,54)
# Select the first 6 features from input_data
input_data = input_data[:6]
input_data_as_numpy_array = np.asarray(input_data)
#reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1) #taking only one datapoint

#Standerizing the data
std_data = scaler.transform(input_data_reshaped)
print(std_data)
prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0] == 0):
 print('NOT DIABETIS')
else:
  print('DIABETIS')
```
